```python
# here is a Python list:

a = [1, 2, 3, 4, 5, 6]
```

```python
# get a list containing the last 3 elements of a
# Yes, you can just type out [4, 5, 6] but we really want to see you demonstrate you know how to do that in Python
a_last3 = []
for x in reversed(range(3)):
    a_last3.append(a[-1 - x])
print(a_last3)
```

```python
# create a list of numbers from 1 to 100
b = []
for x in range(100):
    b.append(x + 1)
print(b)
```

```python
# now get a list with only the even numbers between 1 and 100
# you may or may not make use of the list you made in the last cell
c = []
for x in b:
    if (x % 2 == 0):
        c.append(x)
print(c)
```

```python
# write a function that takes two numbers as arguments
# and returns the first number divided by the second
def division(x, y):
    return x / y
division(6, 2)
```

```python
# write a function that takes a string as input
# and return that string in all caps
def allCaps(string):
    return string.upper()
allCaps("capitalize me")
```

```python
# optional challenge - fizzbuzz
# you will need to use both iteration and control flow 
# go through all numbers from 1 to 100 in order
# if the number is a multiple of 3, print fizz
# if the number is a multiple of 5, print buzz
# if the number is a multiple of 3 and 5, print fizzbuzz and NOTHING ELSE
# if the number is neither a multiple of 3 nor a multiple of 5, print the number
for x in b:
    if (x % 3 == 0 and x % 5 != 0):
        print("fizz\n")
    elif (x % 3 != 0 and x % 5 == 0):
        print("buzz\n")
    elif (x % 3 == 0 and x % 5 == 0):
        print("fizzbuzz\n")
    elif (x % 3 != 0 and x % 5 != 0):
        print(str(x) + "\n")
```

```python
# load in the "starbucks.csv" dataset
# refer to how we read the cereal.csv dataset in the tutorial
df = pd.read_csv("../data/starbucks.csv")
print(df)
```

```python
# output the calories, sugars, and protein columns only of every 40th row.
df.iloc[::40][["calories", "sugars", "protein"]]
```

```python
# select all rows with more than and including 400 calories
df[df["calories"] >= 400]
```

```python
# select all rows whose vitamin c content is higher than the iron content
df[df["vitamin c"] > df["iron"]]
```

```python
# create a new column containing the caffeine per calories of each drink
df["caffeine_per_calories"] = df["caffeine"] / df["calories"]
df["caffeine_per_calories"]
```

```python
# what is the average calorie across all items?
df["calories"].mean()
```

```python
# how many different categories of beverages are there?
print(df["beverage_category"].nunique())
print(df["beverage_category"].unique())
```

```python
# what is the average # calories for each beverage category?
beverage_categories = df.groupby("beverage_category")
beverage_categories["calories"].mean()
```

```python
# plot the distribution of the number of calories in drinks with a histogram
df["calories"].plot.hist(edgecolor='black', alpha=0.8, title="Drink Calories Distribution")
```

```python
# plot calories against total fat with a scatterplot
df.plot.scatter(x="calories", y="total fat", title="Calories vs. Fat Content")
```
